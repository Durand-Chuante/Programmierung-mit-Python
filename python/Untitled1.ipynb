```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]
z = [1, 8, 27, 64, 75]

plt.figure()
plt.subplot(1, 2, 1)
plt.plot(x, y, c = "red", lw = 7, ls = "--", label = "carré")
plt.plot(x, z, c = "blue", lw = 7, ls = ":", label = "cube")
plt.title("Fonction carré et cubique")
plt.xlabel("Valeur")
plt.ylabel("carré ou cube")
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(x, y, c = "blue", lw = 7, ls = "--", label = "carré")
plt.plot(x, z, c = "red", lw = 7, ls = ":", label = "cube")
plt.title("Fonction carré et cubique")
plt.xlabel("Valeur")
plt.ylabel("carré ou cube")
plt.legend()
plt.savefig("Cubique et carré.png")
```
